## Loading Datset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(42)

# generate random data
area = np.random.uniform(low=500, high=2000, size=300)
noise = np.random.normal(loc=0, scale=2000, size=300)
price = 5000 + 10*area + noise

# create dataframe

# Create dataframe
df = pd.DataFrame({'area': area, 'price': price})
df
```

## Summarize Dataset

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

## Visualize Dataset

```python
plt.xlabel("Area")
plt.ylabel("Price")
plt.scatter(df.area,df.price,color="red",marker="*")
```

## Segregate Dataset into Input X and Output Y

```python
X = df.iloc[:,:-1].values
X
```

```python
Y = df.price
Y
```

## Training Dataset Using Linear Regression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

## Splitting Into Train and Test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
X_train
```

```python
model.fit(X_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

## Predicting For Test Data

```python
y_pred = model.predict(X_test)
y_pred
```

## Evaluating Performance

```python
# Compute R-squared value
from sklearn.metrics import r2_score

# Calculate the RÂ² value between the predicted and actual target values
r2 = r2_score(y_test, y_pred)
print('RÂ²:', r2)
```
